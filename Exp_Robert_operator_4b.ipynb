```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/ganesh (1).jpeg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded correctly
if image is None:
    print("Error: Image not found!")
else:
    # Define the Roberts Cross operator kernels
    roberts_x = np.array([[1, 0], [0, -1]], dtype=np.float32)  # Gradient in the x-direction
    roberts_y = np.array([[0, 1], [-1, 0]], dtype=np.float32)  # Gradient in the y-direction

    # Apply the Roberts operator using convolution
    grad_x = cv2.filter2D(image, -1, roberts_x)
    grad_y = cv2.filter2D(image, -1, roberts_y)

    # Compute the gradient magnitude (edge strength)
    grad_magnitude = np.sqrt(grad_x**2 + grad_y**2)

    # Convert the gradient magnitude to uint8 for visualization
    grad_magnitude = np.uint8(np.clip(grad_magnitude, 0, 255))

    # Plotting the results
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 4, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(grad_x, cmap='gray')
    plt.title('Gradient X (Roberts X)')
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(grad_y, cmap='gray')
    plt.title('Gradient Y (Roberts Y)')
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(grad_magnitude, cmap='gray')
    plt.title('Gradient Magnitude (Edge Strength)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
