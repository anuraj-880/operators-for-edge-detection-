```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/ganesh (1).jpeg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded correctly
if image is None:
    print("Error: Image not found!")
else:
    # Prewitt Operator kernels for detecting edges
    prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=np.float32)  # Gradient in the X direction
    prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]], dtype=np.float32)  # Gradient in the Y direction

    # Apply Prewitt operator using convolution
    grad_x = cv2.filter2D(image, -1, prewitt_x)  # Apply Prewitt in X direction
    grad_y = cv2.filter2D(image, -1, prewitt_y)  # Apply Prewitt in Y direction

    # Compute the gradient magnitude (edge strength)
    grad_magnitude = np.sqrt(grad_x**2 + grad_y**2)
    grad_magnitude = np.uint8(np.clip(grad_magnitude, 0, 255))  # Convert to uint8 for visualization

    # Plotting the results
    plt.figure(figsize=(12, 6))

    # Original Image
    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Prewitt Gradient X
    plt.subplot(1, 3, 2)
    plt.imshow(grad_x, cmap='gray')
    plt.title('Prewitt Gradient X')
    plt.axis('off')

    # Prewitt Gradient Y
    plt.subplot(1, 3, 3)
    plt.imshow(grad_y, cmap='gray')
    plt.title('Prewitt Gradient Y')
    plt.axis('off')

    # Gradient Magnitude (Edge Strength)
    plt.figure(figsize=(6, 6))
    plt.imshow(grad_magnitude, cmap='gray')
    plt.title('Prewitt Gradient Magnitude')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
