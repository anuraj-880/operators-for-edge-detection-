```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/ganesh (1).jpeg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded correctly
if image is None:
    print("Error: Image not found!")
else:
    # Gradient in the X direction (using finite differences)
    grad_x = np.diff(image, axis=1, append=image[:, -1].reshape(-1, 1))

    # Gradient in the Y direction (using finite differences)
    grad_y = np.diff(image, axis=0, append=image[-1, :].reshape(1, -1))

    # Compute gradient magnitude (edge strength)
    grad_magnitude = np.sqrt(grad_x**2 + grad_y**2)

    # Plotting the results
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 4, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(grad_x, cmap='gray')
    plt.title('Gradient X (Horizontal Edges)')
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(grad_y, cmap='gray')
    plt.title('Gradient Y (Vertical Edges)')
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(grad_magnitude, cmap='gray')
    plt.title('Gradient Magnitude (Edge Strength)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
